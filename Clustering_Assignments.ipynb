## **K-Means**

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
crime=pd.read_csv('crime_data.csv')
```

```python
crime
```

```python
def minmax_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return (x)
```

```python
df_norm=minmax_func(crime.iloc[:,1:])
df_norm
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime = scaler.fit_transform(crime.iloc[:,1:])
scaled_crime
```

```python
# In KMeans we use elbow chart to find optimum no of clusters
```

```python
wcss = []
for i in range (1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(df_norm)
  wcss.append(kmeans.inertia_)

  print(wcss)

plt.plot(range(1,11),wcss)
plt.title('Elbow chart')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
clusters=KMeans(4,random_state=0)
clusters.fit(df_norm)
```

```python
clusters.labels_
```

```python
crime['clusters']=clusters.labels_
crime
```

```python
crime.groupby('clusters').mean()
```

```python
crime['clusters'].value_counts()
```

```python
import seaborn as sns
sns.scatterplot(data=crime)
```

# **DBSCAN**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
data=crime.values
data
```

```python
scaler = StandardScaler().fit(data)
X = scaler.transform(data)
X
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(data)
x
```

```python
dbscan=DBSCAN(eps=2, min_samples=6)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
crime['Cluster']=dbscan.labels_
```

```python
crime
```

```python
crime['Cluster'].value_counts()
```

```python
x=metrics.silhouette_score(x,dbscan.labels_)
x
```
